```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/fbarth/ml-espm/master/dataset/breast-cancer-wisconsin.csv")
df = df.drop(columns=['id', 'Unnamed: 32'])
df.head()
```

# Beginning of Data Analysis

### -- Mean Radius

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

sns.set_theme(style="dark")
sns.set_palette("colorblind")

sns.histplot(
    data=df,
    x="radius_mean", hue="diagnosis",
    multiple="stack",
    edgecolor=".3",
    linewidth=.5,
    log_scale=False,
)

# mean radius over 17.5 usually means malignant cancer
```

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

sns.set_theme(style="dark")
sns.set_palette("colorblind")

sns.histplot(
    data=df,
    x="smoothness_mean", hue="diagnosis",
    multiple="stack",
    edgecolor=".3",
    linewidth=.5,
    log_scale=False,
)
```

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

sns.set_theme(style="dark")
sns.set_palette("colorblind")

sns.scatterplot(
    data=df, 
    x="radius_mean", 
    y="smoothness_mean", 
    hue="diagnosis")
```
